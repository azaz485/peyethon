# Introduction to Seaborn
- top of matplotlib

- one of the favorite libraries of the author of the course

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(5)
```

```python
sns.get_dataset_names() #to see the available example datasets. 
```

```python
sns.displot(tips['total_bill'], kde= False, bins= 30)
```

```python
sns.histplot(tips['total_bill'],bins=25,kde=True)
```

## Joint Plot
Scatter Plot + Histogram 

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde')
```

## Pair Plot
For every single possible combinations.
Diagonal plots represents the distribution of each variables, x represents the range of values, y (height) represents the count (freq)

```python
sns.pairplot(tips,hue='sex', palette='coolwarm')
```

## Rug Plot
Draws a dashmark for everypoint, for univariate distribution

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'], kde='false') #kernal density estimation plot = kde
```
