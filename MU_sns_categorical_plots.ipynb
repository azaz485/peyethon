# Categorical Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

### Barplot
estimator, applies a function to the total bill variables. Thus, data becomes std(total_bill)

```python
# estimator, applies a function to the total bill variables. Thus, data becomes std(total_bill)
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) #shows average/mean for female and males/
```

### Countplot
similar to bar plot but y axis is already chosen, which is a counter. 

```python
sns.countplot(x='sex',data=tips) 
```

### boxplot and violinplot
these plots are used to show the categorical distribution.
Hue is the strong powerful tool of Seaborn

```python
sns.boxplot(x='day',y='total_bill',data=tips)  
#adding hue splits the plot even more.
# sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

### violin plot
Violin plot showing little more information but they may be hidden. We may need more time to understand violin plot rather than box plot.
If we are representing data to CEO, we can use BarChart. If we can working with data sciencist etc. we can use Violin plot.  

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex',split=True)
```

### Stripplot and swarmplot

Stripplot: scatterplot with categorical data. 
Cons: hard to tell how many data top of each other. 

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex')
```

nice for exploratory Data analysis:

```python
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
#combination of violin and scatter plot. Over each other datas just arranged to places where they can not touch. 
sns.violinplot(x='day',y='total_bill',data=tips)
```

### Factor/CatPlot
Change the kind = '' , and call the all the categorical plots that we see in this lecture. Author says that, he personally prefer calling the plot method itself rather than catplot()

```python
# sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```
