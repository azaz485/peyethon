```python
#import seaborn and inline matploblib, then load the sns.datasets tips and flights
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights') #load the flights dataset
tips.head()
```

```python
flights.head(6)
```

```python
#first we need to chose the numeric columns but using select_dtypes(include='number'
# numeric_columns = tips.select_dtypes(include='number').columns
# correlation = tips[numeric_columns].corr()
# correlation
tc = tips.corr(numeric_only=True)
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm') #set the numbers of correlation values to the heatmap
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers') #create a pivot table to the flights dataset
fp
```

```python
sns.heatmap(fp, cmap='magma') 
```

```python
sns.heatmap(fp, cmap='coolwarm', linecolor='black', linewidths=1) #add lines to the heatmap
```

```python
#clustermap will cluster the rows and columns together based on their similarities
sns.clustermap(fp,figsize=(8,8),cmap='coolwarm')
```

```python
sns.clustermap(fp,figsize=(8,8),cmap='coolwarm',standard_scale=1) #standard_scale=1 will normalize the data
#what does happen when we standardize the data?
#the data will be normalized, so the values will be between 0 and 1
#why we need to normalize the data?
#because the data will be more clear and easy to understand
#notice that year order is changed. 
```
