### Regression Plots and Style and Color

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips, hue='sex',markers=['o','v'],scatter_kws={'s':50}) 
#lmplot is a linear model plot. It is a scatter plot with a linear fit on top of it
#scatter plot with linear fit
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips,col='day', hue='sex',aspect=1) #col and row are used to split the plot into multiple plots
```
